```python
import numpy as np
import matplotlib.pylab as py
a = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
b = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
model=np.poly1d(np.polyfit(a,b,3))
data=np.linspace(1,22,100)
py.plot(data,model(data),label='predicted value',color='red')
py.scatter(a,b,label='my data')
py.xlabel("x")
py.ylabel("y")
py.title("polynomial_regression")

py.legend()

```
